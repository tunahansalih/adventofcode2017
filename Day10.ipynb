```python
import numpy as np
my_input = np.loadtxt('day10input.txt', dtype=np.int, delimiter=',')
```

```python
skip_size = 0
current_index = 0

```

```python
my_list = np.array(range(256))
```

```python

    
for l in my_input:
    my_sublist = np.array([])
    for i in range(current_index, current_index + l):
        my_sublist = np.append([my_list[i%len(my_list)]], my_sublist)
    for i in range(current_index, current_index + l):
        my_list[i%len(my_list)] = my_sublist[i-current_index]
    current_index += l + skip_size
    skip_size += 1
    print(my_list)
```
